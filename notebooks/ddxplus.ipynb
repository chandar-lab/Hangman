```python
import json
import pandas as pd
import numpy as np
```

```python
conditions = pd.read_json('../ddxplus/release_conditions.json').transpose()
evidences = pd.read_json('../ddxplus/release_evidences.json').transpose()

conditions = conditions[['condition_name', 'symptoms', 'antecedents']]
evidences = evidences[evidences.data_type == "B"]
evidences = evidences[['code_question', 'question_en']]

def filter_symptoms(row):
    symptoms_list = list(row['symptoms'].keys())
    symptoms_list = [s for s in symptoms_list if s in evidences.code_question.unique()]
    return symptoms_list

def filter_antecedents(row):
    antecedents_list = list(row['antecedents'].keys())
    antecedents_list = [a for a in antecedents_list if a in evidences.code_question.unique()]
    return antecedents_list

def union_symptoms_and_antecedents(row):
    return list(set(row['symptoms_list'] + row['antecedents_list']))

conditions['symptoms_list'] = conditions.apply(filter_symptoms, axis=1)
conditions['antecedents_list'] = conditions.apply(filter_antecedents, axis=1)
conditions['related'] = conditions.apply(union_symptoms_and_antecedents, axis=1)

conditions = conditions[['condition_name', 'related']]
```

```python
evidences.to_csv('../ddxplus/evidences.csv', index=False)
conditions.to_csv('../ddxplus/conditions.csv', index=False)
```
